# Basic charts
## Load libraries

```python
import plotly.express as px
```

## Scatter Plots

### General scatterplot framework

```python
fig = px.scatter(x=[0, 1, 2, 3, 4], y=[0, 1, 4, 9, 16])
fig.show()
```

### Scatterplot using a dataframe

```python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

### Add another hover variable

```python
fig = px.scatter(
    data_frame=px.data.iris(),
    x="sepal_width",
    y="sepal_length",
    color="species",
    size="petal_length",
    hover_data=["petal_width"]
)
fig.show()
```

### Colors based on a continuous variable

```python
fig = px.scatter(
    data_frame=px.data.iris(),
    x="sepal_width",
    y="sepal_length",
    color="petal_length"
    )
fig.show()
```

### Add a plotting symbol

```python
fig = px.scatter(
    data_frame=px.data.iris(),
    x="sepal_width",
    y="sepal_length",
    color="species",
    symbol="species"
)
fig.show()
```

### Scatter plots and Categorical Axes

```python
df = px.data.medals_long()
fig = px.scatter(
    data_frame = df,
    y = 'nation',
    x = 'count',
    color = 'medal',
    symbol = 'medal'
)
fig.show()
```

### Grouped Scatter Points

```python
from matplotlib.pyplot import scatter
fig = px.scatter(
    data_frame = df,
    y = 'count',
    x = 'nation',
    color = 'medal'
)
fig.update_traces(marker_size = 10)
fig.update_layout(scattermode = 'group')
fig.show()
```

### Control the width of the group

```python
from matplotlib.pyplot import scatter
fig = px.scatter(
    data_frame = df,
    y = 'count',
    x = 'nation',
    color = 'medal'
)
fig.update_traces(marker_size = 10)
fig.update_layout(scattermode = 'group', scattergap = 0.75)
fig.show()
```
