# Violin plots in Python
## Libraries

```python
import plotly.express as px
import polars as pl
import numpy as np
```

## Basic violin plot with plotly express

```python
df = pl.from_pandas(px.data.tips())
fig = px.violin(
    data_frame=df,
    y="total_bill",
)
fig.show()
```

## Violin plot with box and data points

```python
fig = px.violin(
    data_frame=df,
    y="total_bill",
    box=True,
    points="all"
)

fig.show()
```

## Multiple violin plots

```python
fig = px.violin(
    data_frame=df,
    y="tip",
    x="smoker",
    color="sex",
    box=True,
    points="all",
    hover_data=df.columns
)
fig.show()
```

```python
fig = px.violin(
    data_frame=df,
    y="tip",
    color="sex",
    violinmode="overlay",
    hover_data=df.columns
)
fig.show()
```
