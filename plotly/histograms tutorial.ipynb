# Histograms in Python

```python
import plotly.express as px
import polars as pl
import numpy as np
```

## Histograms with Plotly Express
### Example

```python
df = pl.from_pandas(px.data.tips())

fig = px.histogram(
    data_frame = df,
    x="total_bill",
    title="Histogram of Total Bill",
)

fig.show()
```

```python
fig = px.histogram(
    data_frame = df,
    x="day",
    title="Histogram of Days",
)

fig.show()
```

### Choosing the number of bins

```python
fig = px.histogram(
    data_frame = df,
    x="total_bill",
    nbins=20,
    title="Histogram of Total Bill",
)

fig.show()
```

### Histograms on date data

```python
df = pl.from_pandas(px.data.stocks())

fig = px.histogram(
    data_frame = df,
    x="date",
    title="Histogram of Dates",
)
fig.update_layout(bargap=0.2)

fig.show()
```

### Histograms on categorical data

```python
df = pl.from_pandas(px.data.tips())

fig = px.histogram(
    data_frame=df,
    x="day",
    category_orders={"day": ["Thur", "Fri", "Sat", "Sun"]},
    title="Histogram of Days",
)

fig.show()
```
