# Leafmap intro
## Libraries

```python
import leafmap as leafmap
```

## Create an interactive map

```python
m = leafmap.Map()
m
```

## Specify the default map center and zoom level

```python
m = leafmap.Map(center=[40, -119], zoom=4) # the center is (lat, lon)
m
```

## Set the visibility map controls

```python
m = leafmap.Map(
    draw_control=False,
    measure_control=False,
    fullscreen_control=False,
    attribution_control=False
)
m
```

## Change the map width and height

```python
m = leafmap.Map(
    height="450px",
    width="800px"
)
m
```

## Use the ipyleaflet plotting backend

```python
import leafmap.leafmap as leafmap

m = leafmap.Map()
m
```

## Use the folium plotting backend

```python
import leafmap.foliumap as leafmap

m = leafmap.Map()
m
```
